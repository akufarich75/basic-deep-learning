# Pre Processing Image

```
import matplotlib.pyplot as plt
from keras.preprocessing import image

img = image.load_img('/content/1Odq43G.jpg', target_size=(400, 400), color_mode='rgb')

plt.imshow(img)

x = image.img_to_array(img)
print('array sebelum dinormalisasi :')
print(x)
print('')

x2 = x/255
print('array setelah dinormalisasi :')
print(x2)
```

# Pre processing with Keras imagedatagenerator

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib.pyplot import imread, imshow, subplots, show


def plot(data_generator):

  data_generator.fit(images)
  image_iterator = data_generator.flow(images)

  fig, rows = subplots(nrows = 1, ncols=4, figsize=(18,18))
  for row in rows:
    row.imshow(image_iterator.next()[0].astype('int'))
    row.axis('off')
  show

image = imread('/content/1Odq43G.jpg')

images = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))

imshow(images[0])
show()
```

## Rotation

```
data = ImageDataGenerator(rotation_range=90)
plot(data)
```

## Width Shifting

```
width_shifting = ImageDataGenerator(width_shift_range=0.6)
plot(width_shifting)
```

## Height Shifting

```
height_shifting = ImageDataGenerator(height_shift_range=0.3)
plot(height_shifting)
```

## Brightness

```
brightness = ImageDataGenerator(brightness_range=(0.1, 0.8))
plot(brightness)
```

## Zoom Image

```
zoom = ImageDataGenerator(zoom_range=[0.5, 0.7])
plot(zoom)
```

## Horizontal Flip Image

```
horizontal_flip = ImageDataGenerator(horizontal_flip=True)
plot(horizontal_flip)
```

## Vertical Flip Images

```
vertical_flip = ImageDataGenerator(vertical_flip=True)
plot(vertical_flip)
```

## Merubah lebih dari 1 parameter ImageDataGenerator

```
data_baru = ImageDataGenerator(horizontal_flip=True,
                               vertical_flip=True)

plot(data_baru)
```
